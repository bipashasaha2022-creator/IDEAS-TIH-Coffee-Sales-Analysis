```python
df_display =pd.read_csv("Coffe_sales.csv")

df_model = df_display[['hour_of_day', 'coffee_name', 'Weekday', 'money']]
df_model = df_model.dropna()
```

```python
df_encoded = pd.get_dummies(df_model, columns=['coffee_name', 'Weekday'])
df_encoded.head()
```

```python
X = df_encoded.drop("money", axis=1)
y = df_encoded["money"]

model = LinearRegression()
model.fit(X, y)

print("Linear Regression Model Trained on Full Dataset")
```

```python
y_pred=model.predict(X)

mse = mean_squared_error(y, y_pred)
print(" Mean Squared Error (MSE):", mse)
```

```python
#Example of Prediction

sample = pd.DataFrame({
    'hour_of_day': [9],
    'coffee_name': ['Cappuccino'],
    'Weekday': ['Fri']
})

sample_encoded = pd.get_dummies(sample)

sample_encoded = sample_encoded.reindex(columns=X.columns, fill_value=0)

predicted_money = model.predict(sample_encoded)
print("Predicted Sales Amount:", predicted_money[0])
```
